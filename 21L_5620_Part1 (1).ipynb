# ***Question 1***

```python
import re
import nltk
from pprint import pprint
str= 'NaWaLl AaMer %^&*!@#$235.176isadataSCIENCEstuDent CooLPErSoNpot'
regexpat=[
    #Identifies consecutive Upper or Lower case characters
    r'[a-zA-Z]+',
    #Identifies Single Upper case followed by 0 or more lower case characters
    r'[A-Z][a-z]*',
    #Identifies starting character p followed by 0 or less than 2 (0 <= aeiou >2) vowels and word ending in t
    r'p[aeiou]{,2}t',
    #Identifies a decimal character with . and floating points as optional
    r'\d+(\.\d+)?',
    #Identifies that first and last character should be consonants but last should be consonant with a vowel enclosed between them this pattern can be repeated 0 or more times as denoted by*
    r'([^aeiou][aeiou][^aeiou])*',
    #Tokenizes words and alphaneumeric characters ignoring non word and white space characters
    r'\w+|[^\w\s]+',
]
for p in regexpat:
    print(p)
    nltk.re_show(p, str)
    print("\n")
```

# ***Question 3, 4 & 5***

```python
import re
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem import PorterStemmer, LancasterStemmer
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

url_to_scrape = 'https://www.csail.mit.edu/people/?roleFacets=Principal%20Investigators,Core%2FDual,Associates,Emeritus,Core%2FDual,Associates,Emeritus,Principal%20Investigators,Core%2FDual,Associates,Emeritus'
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_service = ChromeService(executable_path='chromedriver.exe')
driver = webdriver.Chrome(service=chrome_service, options=chrome_options)
driver.get(url_to_scrape)
driver.implicitly_wait(10)
page_source = driver.page_source
driver.quit()
soup = BeautifulSoup(page_source, 'html.parser')
result_text = soup.get_text()
if result_text:
    print(result_text)
phone_number_regex = r'\b(?:\d{3}[-.\s]?)?\d{3}[-.\s]?\d{4}\b'
email_regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
phone_numbers = re.findall(phone_number_regex, result_text)
emails = re.findall(email_regex, result_text)
if phone_numbers:
    print("Phone numbers:")
    for phone_number in phone_numbers:
        print(phone_number)
if emails:
    print("\nEmail addresses:")
    for email in emails:
        print(email)
tokenizer = RegexpTokenizer(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b|\b\w+\b')
tokens = tokenizer.tokenize(result_text)
porter_stemmer = PorterStemmer()
lancaster_stemmer = LancasterStemmer()
porter_stems = [porter_stemmer.stem(token) for token in tokens]
lancaster_stems = [lancaster_stemmer.stem(token) for token in tokens]
if tokens:
    print("\nPorter stems:")
    print(porter_stems)
    print("\nLancaster stems:")
    print(lancaster_stems)
```
